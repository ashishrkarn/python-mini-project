### Handling Null Values

```python
# Check if mode is available before using it
insurance_mode =  df['Insurance Provider'].mode()

if not insurance_mode.empty:
    mode_val = insurance_mode[0]
    print(f"Filling nulls in 'Insurance Provider' with mode: {mode_val}")
    df['Insurance Provider'] = df['Insurance Provider'].fillna(mode_val)
else:
    print("No mode found — the column might be entirely null!")
```

**Handling Missing Values in `Blood Type`**

The `Blood Type` column has missing values for approximately 5% of the records.  
Since this feature represents a medical attribute and not a numeric measure, imputing with the **mode** could distort the natural distribution of blood groups.

Instead, we fill missing values with a new category **'Unknown'** — this approach preserves the meaning that the blood type information was unavailable or not recorded.

```python
# Fill Blood Type
df['Blood Type'] = df['Blood Type'].fillna('Unknown')
```

**Context-Aware Imputation for `Billing Amount`**

Instead of filling missing values in `Billing Amount` with a simple mean or median,  
a more reliable approach in healthcare billing data is to consider **contextual factors** that influence hospital costs.  

The following features were identified as strong predictors of billing variation:
- **Medical Condition:** Different illnesses have varying treatment costs.
- **Hospital:** Charges can differ by hospital.
- **Number of Days Admitted:** Longer stays lead to higher bills.
- **Test Results:** Positive/negative outcomes can affect additional procedures.
- **Insurance Provider:** Billing structures vary based on insurance agreements.

We first computed the number of days admitted using the difference between `Discharge Date` and `Date of Admission`, then filled missing billing amounts using the **mean billing for records with the same combination of these factors**.

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
df['Stay_Duration'] = (df['Discharge Date'] - df['Date of Admission']).dt.days

# Group-based mean imputation
group_means = df.groupby(
    ['Medical Condition', 'Stay_Duration', 'Hospital', 'Test Results', 'Insurance Provider']
)['Billing Amount'].transform('mean')

df['Billing Amount'] = df['Billing Amount'].fillna(group_means)
df['Billing Amount'].fillna(df['Billing Amount'].mean(), inplace=True)
```

### Check For Duplicates

```python
df.duplicated().sum()
```

```python
# View the duplicate entries (first 5)
df[df.duplicated()].shape
```

```python
# Step 1: Find all duplicated rows (including originals)
dups_df = df[df.duplicated(keep=False)].copy()

# Step 2: Create a unique "signature" of each row to group identical ones
dups_df['row_signature'] = dups_df.apply(lambda row: tuple(row), axis=1)

# Step 3: Map each signature to its first (original) index
signature_to_original_index = {
    sig: dups_df[dups_df['row_signature'] == sig].index.min()
    for sig in dups_df['row_signature'].unique()
}

# Step 4: Print results in requested format
for idx in dups_df.index:
    sig = dups_df.loc[idx, 'row_signature']
    original_idx = signature_to_original_index[sig]
    
    if idx == original_idx:
        print(f"{idx}. Original row")
    else:
        print(f"{idx}. Duplicate of original row {original_idx}")
```

## Advanced Python Techniques

### Lambda Functions

```python
df['Age_Group'] = df['Age'].apply(lambda x: 'Young' if x < 30 else 'Middle-aged' if 30 <= x < 60 else 'Senior')
df['Age_Group'].value_counts()
```

The age group classification divides the dataset into three categories: "Young" (under 30), "Middle-aged" (30 to 59), and "Senior" (60 and above).

Most patients are in the "Middle-aged" group (26,282), followed by "Senior" (22,471), and there are fewer "Young" patients (10,324).

This distribution indicates the data is skewed toward adults and older individuals, which is typical for many healthcare datasets focused on chronic conditions.

```python
avg_stay = df['Stay_Duration'].mean()
df['Long_Stay'] = df['Stay_Duration'].apply(lambda x: 'Yes' if x > avg_stay else 'No')
df['Long_Stay'].value_counts()
```

The classification of “Long Stay” is based on comparing each patient’s stay duration to the dataset’s average.

Here, the data shows that roughly half the patients (29,613) had shorter-than-average stays, while the other half (29,464) stayed longer.

This balanced split reflects a normal distribution of hospitalization periods, suggesting consistent hospital management for patient discharge timelines.

```python
df['High_Bill'] = df['Billing_Amount'].apply(lambda x: 1 if x > 30000 else 0)
df['High_Bill'].value_counts()
```

The new column “High_Bill” classifies patients based on billing amounts — those with bills above ₹30,000 are labeled as 1 (high), and the rest as 0 (normal).

The results show that 35,136 patients (about 59%) fall under the normal billing category, while 23,941 patients (approximately 41%) have higher-than-average medical bills.

This distinction helps in identifying patients with elevated healthcare costs for further cost-based or insurance-related analysis.
